```python
import cv2

def read_video(video_path):
    print(f"Trying to open video: {video_path}")

    cap = cv2.VideoCapture(video_path)
    cap.set(cv2.CAP_PROP_FOURCC, cv2.VideoWriter_fourcc(*"MJPG"))

    if not cap.isOpened():
        print(f"Could not open video at {video_path}")
        return None

    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

    print(f"Loaded video: {frame_count} frames, {frame_width}x{frame_height}")

    return cap, frame_count, frame_width, frame_height
```

```python
# Function to load .mat zone data
def load_mat_file(mat_path):
    mat_data = scipy.io.loadmat(mat_path)
    
    zones = mat_data['zones']
    
    return zones
```

```python
# Function to extract frame from video
def get_frame(video_cap, frame_number):
    video_cap.set(cv2.CAP_PROP_POS_FRAMES, frame_number)
    ret, frame = video_cap.read()
    if not ret:
        raise ValueError(f"Error for frame {frame_number}.")
    
    # Convert from BGR to RGB
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    
    return frame_rgb
```

```python
def display_sample_frames(video_path, num_samples=5):
    cap, frame_count, frame_width, frame_height = read_video(video_path)
    if cap is None:
        return

    sample_indices = [i * frame_count // num_samples for i in range(num_samples)]
    frames = [get_frame(cap, idx) for idx in sample_indices]
    
    cap.release()

    # Plot the sample frames
    plt.figure(figsize=(15, 5))
    for i, frame in enumerate(frames):
        if frame is not None:
            plt.subplot(1, num_samples, i+1)
            plt.imshow(frame)
            plt.axis("off")
            plt.title(f"Frame {sample_indices[i]}")
    
    plt.show()

# Usage
video_path = "/home/adm808/Sabatini_Lab/Preliminary_Dataset/20220124/NO_1/VID_20130305_233514.avi"
display_sample_frames(video_path)
```

```python
def overlay_zones(frame, zones, alpha=0.2, circle_radius=5):
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.imshow(frame, cmap='gray')

    height, width = frame.shape[:2]  # Get frame dimensions

    for i in range(zones.shape[1]):  
        zone = zones[0, i]
        zone_name = zone["name"][0]
        print(f"Zone name: {zone_name}")

        # Convert `isin` into a 1D array
        isin_values = zone["isin"].flatten()

        # Find pixel indices where `isin == 1`
        active_indices = np.where(isin_values == 1)[0]
        if len(active_indices) == 0:
            print(f"No active pixels found for {zone_name}, skipping.")
            continue

        # # Set maximum for valid indices
        # valid_indices = active_indices[active_indices < len(zone["dim1points"].flatten())]

        # # Extract (x, y) pixel coordinates using `dim1points` and `dim2points`
        y_coords = zone["dim2points"].flatten()[active_indices] - 1  # MATLAB to Python indexing
        x_coords = zone["dim1points"].flatten()[active_indices] - 1

        # row_indices, col_indices = np.unravel_index(active_indices, (height, width))

        # x_coords = np.take(zone["dim1points"].flatten(), row_indices, mode='clip') - 1
        # y_coords = np.take(zone["dim2points"].flatten(), col_indices, mode='clip') - 1

        # Ensure indices do not exceed frame dimensions
        x_coords = np.clip(x_coords, 0, width - 1)
        y_coords = np.clip(y_coords, 0, height - 1)

        print(f"Checking {zone_name}: sum of isin = {np.sum(isin_values)}")
        print(f"Zone {zone_name}: X min/max = {x_coords.min()} / {x_coords.max()}, Y min/max = {y_coords.min()} / {y_coords.max()}")

        # Overlay translucent circles at each coordinate
        for x, y in zip(x_coords, y_coords):
            circle = plt.Circle((x, y), circle_radius, color=(0.7, 0.7, 0.3), alpha=alpha)
            ax.add_patch(circle)

    ax.set_xlim(0, width)
    ax.set_ylim(height, 0)
    ax.axis('off')

    return fig
```

```python
# Main function
def process_video(video_path, mat_path, num_frames=3, output_prefix="frame"):
    cap, frame_count, frame_width, frame_height = read_video(video_path)
    if cap is None:
        return

    zones = load_mat_file(mat_path)

    selected_frames = random.sample(range(frame_count), num_frames)
    
    for frame_num in selected_frames:
        frame = get_frame(cap, frame_num)
        frame = get_frame(cap, frame_num)
        plt.imshow(frame)  
        # Showing the raw frame BEFORE I overlay zones
        plt.title(f"Raw Frame {frame_num}")
        plt.show()
        fig = overlay_zones(frame, zones)

        if fig is None:
            print(f" Skipping frame {frame_num} due to overlay issue.")
            continue  # Skip saving if no valid figure
        
        # Save the image
        output_path = f"{output_prefix}_{frame_num}.png"
        fig.savefig(output_path, bbox_inches='tight')
        plt.close(fig)  # Close to prevent memory issues

    cap.release()
```
