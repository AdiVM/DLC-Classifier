```python
# It will say no labels initially, but after running next secitons labels will be created
# Biggest issue was that the version of the file I saved has an extra column now, which is what is causing the error
# So I fixed that here by just starting at column position 1 instead of 0, basically skipping over that column
```

```python
# This first converts the csv from the raw format I collected into a format with the coordinates in the proper places

import pandas as pd
import os

# Paths
raw_csv = "/n/scratch/users/a/adm808/Sabatini_Lab/ZoneDetection2-adi-2025-04-23/labeled-data/dummy_video/CollectedData_Auto.csv"
fixed_csv = "/n/scratch/users/a/adm808/Sabatini_Lab/ZoneDetection2-adi-2025-04-23/labeled-data/dummy_video/CollectedData_adi_nopaths.csv"

# Load CSV
df = pd.read_csv(raw_csv)

# Strip any leading path from image filenames
df.iloc[:, 1] = df.iloc[:, 1].apply(os.path.basename)

# Split image column and coordinate matrix
image_col = df.iloc[:, 1]
coords = df.iloc[:, 2:]
coords = coords.loc[:, coords.columns.notna()]  # drop any NaN columns just in case
coords = coords.iloc[:, :17]  # hard-trim to 16 coordinate columns (8 zones × 2)

# Build new header
n_zones = coords.shape[1] // 2
scorer_row = ['adi'] * (n_zones * 2)
bodypart_row = [f"Zone{i}" for i in range(1, n_zones + 1) for _ in ('x', 'y')]
coords_row = ['x', 'y'] * n_zones

multiindex = pd.MultiIndex.from_arrays(
    [scorer_row, bodypart_row, coords_row],
    names=["scorer", "bodyparts", "coords"]
)

# Assign new headers
coords.columns = multiindex
coords.insert(0, ('scorer', 'bodyparts', 'coords'), image_col)

# Write to CSV
coords.to_csv(fixed_csv, index=False)
print("CSV reformatted and saved to:", fixed_csv)

# So now our csv is formatted correctly
```

```python
# Lastly, I clean the file so that it only contains the names of images that actually are present in the directory
import pandas as pd
import os

# Project root — the directory with config.yaml
project_path = "/n/scratch/users/a/adm808/Sabatini_Lab/ZoneDetection2-adi-2025-04-23"

# Input and output CSVs
csv_in = f"{project_path}/labeled-data/dummy_video/CollectedData_adi_preclean.csv"
csv_out = f"{project_path}/labeled-data/dummy_video/CollectedData_adi.csv"

# Load CSV with DLC-style 3-row column header
df = pd.read_csv(csv_in, header=[0, 1, 2])

# Step 1: Filter rows where the file in the first column DOES exist
df = df[df.iloc[:, 0].apply(lambda x: os.path.exists(os.path.join(project_path, x)))]

# Step 2: Set the first column (image path) as the index
df.index = df.iloc[:, 0]

# Step 3: Drop the original path column (now redundant since it’s the index)
df.drop(columns=df.columns[0], inplace=True)

df.columns.set_names(["scorer", "bodyparts", "coords"], inplace=True)

# Step 4: Save with correct structure (DLC expects index = image path)
df.to_csv(csv_out)

print("Cleaned and formatted for DeepLabCut")
```
